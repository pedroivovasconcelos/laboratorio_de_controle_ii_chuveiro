```python
import numpy as np
import matplotlib.pyplot as plt
from control.matlab import *

# Parâmetros da planta
omega_n = 10  # frequência natural
zeta = 0.5  # fator de amortecimento

# Criando a função de transferência da planta
num = [omega_n**2]
den = [1, 2*zeta*omega_n, omega_n**2]
planta = tf(num, den)

# Definindo o sinal de entrada e o vetor de tempo
tempo = np.linspace(0, 10, 1000)  # 10 segundos, 1000 pontos
entrada = np.sin(tempo)  # sinal de entrada é um seno

# Gerando a resposta da planta ao sinal de entrada
yout, resposta, xout = lsim(planta, entrada, tempo)

# Plotando a resposta da planta
plt.figure()
plt.plot(tempo, xout)
plt.title('Resposta da planta ao sinal de entrada')
plt.xlabel('Tempo [s]')
plt.ylabel('Resposta da planta')
plt.grid(True)
plt.show()
```
